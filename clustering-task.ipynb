```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.pipeline import Pipeline
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('/kaggle/input/customer-data-clustring/Customer-Data - 2.csv')
df.head()
```

```python
df.drop('CUST_ID', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(), inplace=True)
df.isnull().sum()
```

```python
SS = StandardScaler()
scaled_df = SS.fit_transform(df)
```

```python
kmeans = KMeans(n_clusters = 5, random_state=42)
kmeans.fit(scaled_df)

km_labels = kmeans.predict(scaled_df)

s_score = silhouette_score(scaled_df, km_labels)
print(s_score)
```

```python
# Define the Agglomerative Clustering model
agg_clustering = AgglomerativeClustering(n_clusters=5)

# Train the model and predict cluster labels
labels_agg = agg_clustering.fit_predict(scaled_df)

# Evaluate the model
silhouette_agg = silhouette_score(scaled_df, labels_agg)
print("Agglomerative Clustering Silhouette Score:", silhouette_agg)
```
